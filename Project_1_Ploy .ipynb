### Research Question 1 (which movie genre was the most popular in the lastest year?)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
```

```python
# Find latest year
latest_year = df['release_year'].max()
latest_year
```

```python
#filter the lastest year 
latest_year_df = df[df['release_year'] == latest_year]

latest_year_df[['original_title', 'genres', 'popularity', 'release_year']].head()
```

```python
#split multiple genre in "genres"

genre_data = latest_year_df.assign(
    genres=latest_year_df['genres'].str.split('|')
).explode('genres')
```

```python
#find the most popular genre by the popularity in 2015
genre_popularity = genre_data.groupby('genres')['popularity'].mean().sort_values(ascending=False)

genre_popularity
```

```python
#finding the top genre in 2015 based on the movie popularity 

most_popular_genre = genre_popularity.idxmax()
highest_avg_popularity = genre_popularity.max()

print("The most popular genre in", latest_year, "was", most_popular_genre)
print("Average popularity:", highest_avg_popularity)
```

```python
genre_popularity.plot(kind='bar', figsize=(12,6))

plt.title(f'Average Popularity by Genre in {latest_year}')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45)
plt.show()
```

### Research Question 2  (Does higher budget movie would generate high revenue?)

```python
#next we will find out if higher-budger movie generate more revenue

# keep only movies with valid budget and revenue

budget_revenue_df = df[(df['budget'] > 0) & (df['revenue'] > 0)]

```

```python
#if the number of the correlation is closer to 1, strong positive relationship

correlation = budget_revenue_df['budget'].corr(
    budget_revenue_df['revenue']
)

print("Correlation between budget and revenue:", correlation)
```

```python
budget_revenue_df.plot(
    x='budget',
    y='revenue',
    kind='scatter',
    figsize=(10,6),
    alpha=0.5
)

plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

To answer whether higher-budget movies generate higher revenue, I compared the budget and revenue columns for all movies with non-zero values. A scatter plot was created to visualize the relationship, and the Pearson correlation coefficient was calculated.

The analysis shows a positive correlation between movie budget and revenue, suggesting that movies with larger budgets generally earn higher revenue. However, some lower-budget movies also performed very well financially, meaning budget alone does not determine success.
